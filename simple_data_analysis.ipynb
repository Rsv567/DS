```python
#include <iostream>
#include <vector>
#include <cstdlib>
#include <ctime>

#define üòÉ int
#define üéâ void
#define üçé std::vector<int>
#define üñ®Ô∏è std::cout
#define üî¢ rand()
#define ‚è∞ time(0)
#define üîÄ srand
#define üîÅ for
#define üîÑ while
#define üé≤ if
#define üéà else
#define üöÄ return
#define üéØ main
#define üñ•Ô∏è std::endl

üéâ üé≠(üçé& üçè, üòÉ üçå, üòÉ üçç, üòÉ üçá) {
    üòÉ üçí = üçç - üçå + 1;
    üòÉ üçì = üçá - üçç;

    üçé üçë(üçí), üçê(üçì);

    üîÅ (üòÉ üçî = 0; üçî < üçí; üçî++)
        üçë[üçî] = üçè[üçå + üçî];
    üîÅ (üòÉ üçî = 0; üçî < üçì; üçî++)
        üçê[üçî] = üçè[üçç + 1 + üçî];

    üòÉ üçî = 0, üçï = 0, üçñ = üçå;
    üîÑ (üçî < üçí && üçï < üçì) {
        üé≤ (üçë[üçî] <= üçê[üçï]) {
            üçè[üçñ] = üçë[üçî];
            üçî++;
        } üéà {
            üçè[üçñ] = üçê[üçï];
            üçï++;
        }
        üçñ++;
    }

    üîÑ (üçî < üçí) {
        üçè[üçñ] = üçë[üçî];
        üçî++;
        üçñ++;
    }

    üîÑ (üçï < üçì) {
        üçè[üçñ] = üçê[üçï];
        üçï++;
        üçñ++;
    }
}

üéâ üçø(üçé& üçè, üòÉ üçå, üòÉ üçá) {
    üé≤ (üçå < üçá) {
        üòÉ üçç = üçå + (üçá - üçå) / 2;

        üçø(üçè, üçå, üçç);
        üçø(üçè, üçç + 1, üçá);

        üé≠(üçè, üçå, üçç, üçá);
    }
}

üçé üé≤(üòÉ üéØ, üòÉ üé∞, üòÉ üé≥) {
    üçé üé®(üéØ);
    üîÅ (üòÉ üçî = 0; üçî < üéØ; üçî++) {
        üé®[üçî] = üî¢ % (üé≥ - üé∞ + 1) + üé∞;
    }
    üöÄ üé®;
}

üéâ üñ®Ô∏è(const üçé& üçè) {
    üîÅ (üòÉ üçî : üçè) {
        üñ®Ô∏è << üçî << " ";
    }
    üñ®Ô∏è << üñ•Ô∏è;
}

üòÉ üéØ() {
    üîÄ(‚è∞);

    üòÉ üéØ = 20;
    üòÉ üé∞ = 0;
    üòÉ üé≥ = 100;

    üçé üçè = üé≤(üéØ, üé∞, üé≥);

    üñ®Ô∏è << "?: ";
    üñ®Ô∏è(üçè);

    üçø(üçè, 0, üçè.size() - 1);

    üñ®Ô∏è << "??: ";
    üñ®Ô∏è(üçè);

    üöÄ 0;
}
```

```python
import random

def merge(arr, left, mid, right):
    left_part = arr[left:mid+1]
    right_part = arr[mid+1:right+1]

    i = j = 0
    k = left

    while i < len(left_part) and j < len(right_part):
        if left_part[i] <= right_part[j]:
            arr[k] = left_part[i]
            i += 1
        else:
            arr[k] = right_part[j]
            j += 1
        k += 1

    while i < len(left_part):
        arr[k] = left_part[i]
        i += 1
        k += 1

    while j < len(right_part):
        arr[k] = right_part[j]
        j += 1
        k += 1

def merge_sort(arr, left, right):
    if left < right:
        mid = left + (right - left) // 2
        merge_sort(arr, left, mid)
        merge_sort(arr, mid + 1, right)
        merge(arr, left, mid, right)

def generate_random_list(size, min_val, max_val):
    return [random.randint(min_val, max_val) for _ in range(size)]

def print_list(arr):
    print(" ".join(map(str, arr)))

def main():
    random.seed()
    size = 20
    min_val = 0
    max_val = 100

    arr = generate_random_list(size, min_val, max_val)
    print("?:", end=" ")
    print_list(arr)

    merge_sort(arr, 0, len(arr) - 1)

    print("??:", end=" ")
    print_list(arr)

if __name__ == "__main__":
    main()
```

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –º–æ–¥—É–ª—å –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–º–∏ –æ–±—ä–µ–∫—Ç–∞–º–∏ Plotly
import plotly.graph_objects as go
import numpy as np  # –î–æ–±–∞–≤–∏–º –∏–º–ø–æ—Ä—Ç numpy, —Ç–∞–∫ –∫–∞–∫ np –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ –∫–æ–¥–µ

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è 3D-–≥—Ä–∞—Ñ–∏–∫–∞ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏
def plot_3d_surface():
    # –°–æ–∑–¥–∞–µ–º –º–∞—Å—Å–∏–≤—ã –∑–Ω–∞—á–µ–Ω–∏–π X –∏ Y –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç -10 –¥–æ 10 —Å 200 —à–∞–≥–∞–º–∏
    x = np.linspace(-10, 10, 200)
    y = np.linspace(-10, 10, 200)

    # –°–æ–∑–¥–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—É—é —Å–µ—Ç–∫—É
    x, y = np.meshgrid(x, y)

    # –í—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è Z –Ω–∞ –æ—Å–Ω–æ–≤–µ –∫–æ—Å–∏–Ω—É—Å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
    z = np.cos(np.sqrt(x**2 + y**2))

    # –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç –≥—Ä–∞—Ñ–∏–∫–∞ —Å –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å—é
    fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

    # –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
    fig.update_layout(
        title='Cosine Function 3D Surface Plot',  # –ó–∞–≥–æ–ª–æ–≤–æ–∫ –≥—Ä–∞—Ñ–∏–∫–∞
        scene=dict(
            xaxis_title='X Axis',  # –ü–æ–¥–ø–∏—Å—å –æ—Å–∏ X
            yaxis_title='Y Axis',  # –ü–æ–¥–ø–∏—Å—å –æ—Å–∏ Y
            zaxis_title='Z Axis',  # –ü–æ–¥–ø–∏—Å—å –æ—Å–∏ Z
            aspectmode='cube',  # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Å—Ç–æ—Ä–æ–Ω –≤ –≤–∏–¥–µ –∫—É–±–∞
        ),
        margin=dict(l=0, r=0, b=0, t=0)  # –£–±–∏—Ä–∞–µ–º –≤–Ω–µ—à–Ω–∏–µ –æ—Ç—Å—Ç—É–ø—ã
    )

    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
    fig.show()

# –í—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
plot_3d_surface()
```

3. –í—ã—è—Å–Ω–∏—Ç–µ, –∫–∞–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ –Ω–∞—Å—Ç—Ä–æ–π–∫—É —Ü–≤–µ—Ç–æ–≤–æ–π –ø–∞–ª–∏—Ç—Ä—ã –≥—Ä–∞—Ñ–∏–∫–∞? –ò–∑–º–µ–Ω–∏—Ç–µ –∫–æ–¥ —Ç–∞–∫, —á—Ç–æ–±—ã –ø–∞–ª–∏—Ç—Ä–∞ –±—ã–ª–∞ –æ—Ç —Å–≤–µ—Ç–ª–æ-—Ä–æ–∑–æ–≤–æ–≥–æ –¥–æ —Å–≤–µ—Ç–ª–æ-–≥–æ–ª—É–±–æ–≥–æ.

–í Plotly –ø–∞—Ä–∞–º–µ—Ç—Ä, –æ—Ç–≤–µ—á–∞—é—â–∏–π –∑–∞ –Ω–∞—Å—Ç—Ä–æ–π–∫—É —Ü–≤–µ—Ç–æ–≤–æ–π –ø–∞–ª–∏—Ç—Ä—ã 3D-–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏, ‚Äî —ç—Ç–æ colorscale.

```python
import plotly.graph_objects as go
import numpy as np

def plot_3d_surface():
    x = np.linspace(-10, 10, 200)
    y = np.linspace(-10, 10, 200)
    x, y = np.meshgrid(x, y)
    z = np.cos(np.sqrt(x**2 + y**2))

    # –°–æ–∑–¥–∞–µ–º –∫–∞—Å—Ç–æ–º–Ω—É—é —Ü–≤–µ—Ç–æ–≤—É—é —à–∫–∞–ª—É –æ—Ç —Å–≤–µ—Ç–ª–æ-—Ä–æ–∑–æ–≤–æ–≥–æ –¥–æ —Å–≤–µ—Ç–ª–æ-–≥–æ–ª—É–±–æ–≥–æ
    custom_colorscale = [
        [0, 'rgb(255, 182, 193)'],  # –°–≤–µ—Ç–ª–æ-—Ä–æ–∑–æ–≤—ã–π
        [1, 'rgb(173, 216, 230)']   # –°–≤–µ—Ç–ª–æ-–≥–æ–ª—É–±–æ–π
    ]

    fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale=custom_colorscale)])

    fig.update_layout(
        title='Cosine Function 3D Surface Plot',
        scene=dict(
            xaxis_title='X Axis',
            yaxis_title='Y Axis',
            zaxis_title='Z Axis',
            aspectmode='cube',
        ),
        margin=dict(l=0, r=0, b=0, t=0)
    )

    fig.show()

plot_3d_surface()
```

–î–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏—è –Ω—É–∂–Ω–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –∫–∞–∂–¥–æ–π –∫–æ–º–ø–∞–Ω–∏–∏, –∞ —Ç–∞–∫–∂–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫.

–ó–∞–≥—Ä—É–∑–∏—Ç–µ —Ç–∞–±–ª–∏—Ü—É –æ—Ü–µ–Ω–æ–∫ `yelp_review.csv`.

```python
df_reviews = pd.read_csv('yelp_review.csv')

print(df_reviews.head())
```

–í –ø–æ–¥–≥—Ä—É–∂–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ –æ—Ü–µ–Ω–æ–∫ –æ—Å—Ç–∞–≤—å—Ç–µ —Ç–æ–ª—å–∫–æ –∫–æ–º–ø–∞–Ω–∏–∏ –≥–æ—Ä–æ–¥–∞ `N`. –î–ª—è —ç—Ç–æ–≥–æ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è `business_id` –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–∞ —É —Ç–∞–±–ª–∏—Ü—ã –æ—Ü–µ–Ω–æ–∫ –∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –º–µ—Ç–æ–¥–æ–º `loc`.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞.* –ß—Ç–æ–±—ã –∏–Ω–¥–µ–∫—Å —Å–Ω–æ–≤–∞ —Å–¥–µ–ª–∞—Ç—å –ø–æ–ª–µ–º —Ç–∞–±–ª–∏—Ü—ã, –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –º–µ—Ç–æ–¥–æ–º `reset_index`.

```python
df_reviews = pd.read_csv('yelp_review.csv')
df_business = pd.read_csv('yelp_business.csv')

city_name = 'Las Vegas'
df_business_city_n = df_business[df_business['city'] == city_name]

df_reviews.set_index('business_id', inplace=True)

df_reviews_filtered = df_reviews.loc[df_reviews.index.isin(df_business_city_n['business_id'])]

df_reviews_filtered.reset_index(inplace=True)

print(df_reviews_filtered)
```

–¢–µ–ø–µ—Ä—å –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –∫–∞–∂–¥–æ–π –∫–æ–º–ø–∞–Ω–∏–∏, –∞ —Ç–∞–∫–∂–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –∫–æ–º–ø–∞–Ω–∏–∏ –æ—Ü–µ–Ω–æ–∫.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞.* –ü–æ–º–æ—á—å –≤ —ç—Ç–æ–º –º–æ–≥—É—Ç —Ñ—É–Ω–∫—Ü–∏–∏ `groupby` –∏ `aggregate([np.mean, np.size])`.

```python
df_reviews_aggregated = df_reviews_filtered.groupby('business_id').agg(
    average_rating=('stars', 'mean'),
    review_count=('stars', 'size')
).reset_index()

print(df_reviews_aggregated)
```

–ù–∞–∑–æ–≤–∏—Ç–µ –∫–æ–ª–æ–Ω–∫–∏ —Ç–∞–±–ª–∏—Ü—ã –∫—Ä–∞—Å–∏–≤—ã–º–∏ –∏–º–µ–Ω–∞–º–∏, –∏–∑–º–µ–Ω–∏–≤ `<–∏–º—è —Ç–∞–±–ª–∏—Ü—ã>.columns`, –ø–æ—Å–ª–µ —á–µ–≥–æ –Ω–∞–ø–µ—á–∞—Ç–∞–π—Ç–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã. –ö—Ä–∞—Å–∏–≤—ã–µ –∏–º–µ–Ω–∞ &mdash; —Ç–æ, —á—Ç–æ –±—É–¥–µ—Ç –ø–æ–Ω—è—Ç–Ω–æ –ø—Ä–æ—Å—Ç–æ–º—É —á–µ–ª–æ–≤–µ–∫—É –ø—Ä–∏ —á—Ç–µ–Ω–∏–∏ –≤–∞—à–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤. –ö–∞–∫ –∏–º–µ–Ω–Ω–æ –∏—Ö –Ω–∞–∑–≤–∞—Ç—å &mdash; –∑–∞–¥–∞—á–∞ –∞–Ω–∞–ª–∏—Ç–∏–∫–∞, —Ç–æ –µ—Å—Ç—å –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –≤–∞—à–∞ :)

```python
df_reviews_aggregated = df_reviews_filtered.groupby('business_id').agg(
    average_rating=('stars', 'mean'),
    review_count=('stars', 'size')
).reset_index()

df_reviews_aggregated.columns = ['Business ID', 'Average Rating', 'Number of Reviews']

print(df_reviews_aggregated.head())
```

–°–æ–µ–¥–∏–Ω–∏—Ç–µ –¥–≤–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–∞–Ω–µ–µ —Ç–∞–±–ª–∏—Ü—ã –ø–æ –∫–æ–º–ø–∞–Ω–∏—è–º –≥–æ—Ä–æ–¥–∞ `N` –≤ –æ–¥–Ω—É.
–î–ª—è —ç—Ç–æ–≥–æ —Å–Ω–∞—á–∞–ª–∞ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –ø–æ–ª–µ `business_id` –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–∞ –≤ –æ–±–µ–∏—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö —Å –ø–æ–º–æ—â—å—é `set_index`. –í –æ–¥–Ω–æ–π –∏–∑ –Ω–∏—Ö —ç—Ç–æ —É–∂–µ –¥–æ–ª–∂–Ω–æ –±—ã–ª–æ –±—ã—Ç—å —Å–¥–µ–ª–∞–Ω–æ.
–í –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ –¥–æ–ª–∂–Ω—ã –ø–æ–ª—É—á–∏—Ç—Å—è –ø–æ–ª—è `latitude`, `longitude`, `categories`, `name`, –∞ —Ç–∞–∫–∂–µ –¥–≤–µ –∫–æ–ª–æ–Ω–∫–∏ —Å–æ —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–æ–º–ø–∞–Ω–∏–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –æ—Ü–µ–Ω–æ–∫, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã –ø–æ—Å—á–∏—Ç–∞–ª–∏ –≤—ã—à–µ.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞.* –°–æ–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü –º–æ–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Å –ø–æ–º–æ—â—å—é `join`.
–ò–Ω–¥–µ–∫—Å—ã —É —ç—Ç–∏—Ö —Ç–∞–±–ª–∏—Ü –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ, —Ç–∞–∫ —á—Ç–æ —Ç–∏–ø –¥–∂–æ–π–Ω–∞ –Ω–µ –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è.

```python
df_reviews = pd.read_csv('yelp_review.csv')
df_business = pd.read_csv('yelp_business.csv')

city_name = 'Las Vegas'
df_business_city_n = df_business[df_business['city'] == city_name]

df_business_city_n.set_index('business_id', inplace=True)

df_reviews.set_index('business_id', inplace=True)
df_reviews_filtered = df_reviews.loc[df_reviews.index.isin(df_business_city_n.index)]

df_reviews_aggregated = df_reviews_filtered.groupby('business_id').agg(
    average_rating=('stars', 'mean'),
    number_of_reviews=('stars', 'size')
).reset_index()

df_reviews_aggregated.columns = ['Business ID', 'Average Rating', 'Number of Reviews']

df_reviews_aggregated.set_index('Business ID', inplace=True)

df_combined = df_business_city_n.join(df_reviews_aggregated)

df_combined = df_combined[['latitude', 'longitude', 'categories', 'name', 'Average Rating', 'Number of Reviews']]

print(df_combined.head())
```

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->
<!-- bba9ec349279ff7d3caf3717aabafe3e -->
–ò–∑–æ–±—Ä–∞–∑–∏—Ç–µ –≤—Å–µ –∫–æ–º–ø–∞–Ω–∏–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ, —Ä–∞—Å–∫—Ä–∞—Å–∏–≤ —Ç–æ—á–∫—É –≤ —Ü–≤–µ—Ç, –æ—Ç—Ç–µ–Ω–æ–∫ –∫–æ—Ç–æ—Ä–æ–≥–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–µ –∫–æ–º–ø–∞–Ω–∏–∏. –ü—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å —Ç–æ—á–∫–∏ –≤—ã—Å—Ç–∞–≤–ª—è–π—Ç–µ –Ω–µ –±–æ–ª–µ–µ $0.3$. –ï—Å–ª–∏ —É –æ–¥–Ω–æ–π –∫–æ–º–ø–∞–Ω–∏–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ–∏–ª–∏–∞–ª–æ–≤ —Å —Ä–∞–∑–Ω—ã–º–∏ –∞–¥—Ä–µ—Å–∞–º–∏, –Ω–∞–Ω–æ—Å–∏—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –≤—Å–µ –∞–¥—Ä–µ—Å–∞.

```python
las_vegas_map = folium.Map(location=[36.1699, -115.1398], zoom_start=13, tiles='OpenStreetMap')

marker_cluster = MarkerCluster().add_to(las_vegas_map)

df_filtered = df_combined[df_combined['Number of Reviews'] >= 5]

df_filtered_high_rating = df_filtered[df_filtered['Average Rating'] >= 4.0]

for _, row in df_filtered_high_rating.iterrows():
    popup_content = f"<b>{row['name']}</b><br>–†–µ–π—Ç–∏–Ω–≥: {row['Average Rating']}, –ö–æ–ª-–≤–æ –æ—Ç–∑—ã–≤–æ–≤: {row['Number of Reviews']}"

    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=popup_content,
        icon=folium.Icon(color='green' if row['Average Rating'] >= 4.5 else ('orange' if row['Average Rating'] >= 3.5 else 'red'))
    ).add_to(marker_cluster)

las_vegas_map.save("las_vegas_map_with_ratings_optimized.html")

print("–ö–∞—Ä—Ç–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –∫–∞–∫ 'las_vegas_map_with_ratings_optimized.html'")
```

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->
<!-- 00000000000000000000000000000000 -->
–ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å —Ä–∞–π–æ–Ω—ã –≥–æ—Ä–æ–¥–∞, —Ç–æ –µ—Å—Ç—å —Ä–∞–∑–¥–µ–ª–∏—Ç—å –≥–æ—Ä–æ–¥ –Ω–∞ "–∫–ª–µ—Ç–∫–∏", –æ–∫—Ä—É–≥–ª–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è —à–∏—Ä–æ—Ç—ã –∏ –¥–æ–ª–≥–æ—Ç—ã, –ø–æ–¥–æ–±—Ä–∞–≤ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Ä–∞–π–æ–Ω–∞.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞.* –ù–∞–ø—Ä–∏–º–µ—Ä, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Ç–∞–∫
`np.round(–¥–æ–ª–≥–æ—Ç–∞*4, decimals=1)*0.25`.

```python
city_name = 'Las Vegas'
df_business_city_n = df_business[df_business['city'] == city_name].copy()

df_business_city_n.loc[:, 'latitude_rounded'] = np.round(df_business_city_n['latitude']*4, decimals=1)*0.25
df_business_city_n.loc[:, 'longitude_rounded'] = np.round(df_business_city_n['longitude']*4, decimals=1)*0.25

print(df_business_city_n[['latitude', 'longitude', 'latitude_rounded', 'longitude_rounded']].head())

```

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->
<!-- 228c0524733d5289e94cde00fc9b704b -->
–î–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–∏ –∫–æ–º–ø–∞–Ω–∏–π –ø–æ —Ä–∞–π–æ–Ω—É –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø—Ä–∏ –ø–æ–º–æ—â–∏ `pd.pivot_table`, –≤–∑—è–≤ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–æ–≤ –∏ –∫–æ–ª–æ–Ω–æ–∫ –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—ã–µ —à–∏—Ä–æ—Ç—É –∏ –¥–æ–ª–≥–æ—Ç—É, –∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∑–Ω–∞—á–µ–Ω–∏–π &mdash; –æ—Ü–µ–Ω–∫–∏. –ê–≥–≥—Ä–µ–≥–∏—Ä—É—é—â–µ–π —Ñ—É–Ω–∫—Ü–∏–µ–π —è–≤–ª—è–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–µ.

–ò–∑–æ–±—Ä–∞–∑–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø—Ä–∏ –ø–æ–º–æ—â–∏ `sns.heatmap`.

*–ó–∞–º–µ—á–∞–Ω–∏–µ.* –í –∫–∞—á–µ—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –º–æ–∂–Ω–æ –±—ã–ª–æ –±—ã –ø–æ—Å—á–∏—Ç–∞—Ç—å –∫–∞–∫ –æ–±—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ –≤—Å–µ–º –æ—Ü–µ–Ω–∫–∞–º –≤—Å–µ—Ö –∫–æ–º–ø–∞–Ω–∏–π, —Ç–∞–∫ –∏ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ –∫–æ–º–ø–∞–Ω–∏—è–º –∏—Ö —Å—Ä–µ–¥–Ω–∏—Ö –æ—Ü–µ–Ω–æ–∫. –í —á–µ–º —Ä–∞–∑–Ω–∏—Ü–∞ —ç—Ç–∏—Ö –ø–æ–¥—Ö–æ–¥–æ–≤? –ö–∞–∫–æ–π –∏–∑ –Ω–∏—Ö –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ?

```python
df_reviews_aggregated = df_reviews_filtered.groupby('business_id').agg(
    average_rating=('stars', 'mean')
).reset_index()

df_combined = df_business_city_n.set_index('business_id').join(df_reviews_aggregated.set_index('business_id'))

pivot_table = pd.pivot_table(df_combined,
                             values='average_rating',
                             index='latitude_rounded',
                             columns='longitude_rounded',
                             aggfunc='mean')

plt.figure(figsize=(14, 12))

sns.heatmap(pivot_table,
            cmap="coolwarm",
            annot=True,
            fmt='.2f',
            annot_kws={'size': 9},
            cbar_kws={'label': '–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞'},
            linewidths=0.5,
            linecolor='white')

plt.title('–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –∫–æ–º–ø–∞–Ω–∏–π –ø–æ —Ä–∞–π–æ–Ω–∞–º –õ–∞—Å-–í–µ–≥–∞—Å–∞')
plt.show()
```

–ü–æ–ª—É—á–µ–Ω–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –∏–º–µ–µ—Ç —Ä—è–¥ –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–æ–≤. –í–æ-–ø–µ—Ä–≤—ã—Ö, –Ω–µ –æ—á–µ–Ω—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Å—É–¥–∏—Ç—å –æ —Ä–∞–π–æ–Ω–µ, –µ—Å–ª–∏ –≤ –Ω–µ–º –º–∞–ª–æ –∫–æ–º–ø–∞–Ω–∏–π. –í–æ-–≤—Ç–æ—Ä—ã—Ö,  –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ —Ü–≤–µ—Ç–æ–≤–∞—è –≥–∞–º–º–∞ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –ø–æ–¥—Å—Ç—Ä–æ–∏–ª–∞—Å—å –ø–æ–¥ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ—Ü–µ–Ω–∫–∏.

–ü–æ—á–µ–º—É —ç—Ç–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–∏ –º–æ–≥—É—Ç –±—ã—Ç—å —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–º–∏?

**–û—Ç–≤–µ—Ç:**
1. –ú–∞–ª–æ –∫–æ–º–ø–∞–Ω–∏–π –≤ —Ä–∞–π–æ–Ω–µ:

    –ï—Å–ª–∏ –≤ —Ä–∞–π–æ–Ω–µ –º–∞–ª–æ –∫–æ–º–ø–∞–Ω–∏–π, —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏—Å–∫–∞–∂–µ–Ω–∞ —Å–ª—É—á–∞–π–Ω—ã–º–∏ –∞–Ω–æ–º–∞–ª—å–Ω—ã–º–∏ –æ—Ü–µ–Ω–∫–∞–º–∏, —á—Ç–æ –¥–µ–ª–∞–µ—Ç –≤—ã–≤–æ–¥—ã –Ω–µ–Ω–∞–¥–µ–∂–Ω—ã–º–∏.

2. –¶–≤–µ—Ç–æ–≤–∞—è –≥–∞–º–º–∞ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –ø–æ–¥—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç—Å—è:

    –¶–≤–µ—Ç–æ–≤–∞—è –≥–∞–º–º–∞, –ø–æ–¥—Å—Ç—Ä–∞–∏–≤–∞—é—â–∞—è—Å—è –ø–æ–¥ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –º–æ–∂–µ—Ç –≤—ã–¥–µ–ª–∏—Ç—å —ç–∫—Å—Ç—Ä–µ–º–∞–ª—å–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏, –∏—Å–∫–∞–∂–∞—é—â–∏–µ –≤–æ—Å–ø—Ä–∏—è—Ç–∏–µ –±–æ–ª–µ–µ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Ä–∞–π–æ–Ω–æ–≤.

–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:

- –£—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –ø–æ—Ä–æ–≥ –¥–ª—è —á–∏—Å–ª–∞ –∫–æ–º–ø–∞–Ω–∏–π –≤ —Ä–∞–π–æ–Ω–µ.

- –†–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –≤–∑–≤–µ—à–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –¥–ª—è –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ–π –∫–∞—Ä—Ç–∏–Ω—ã.

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->

–û—Å—Ç–∞–≤—å—Ç–µ —Ä–∞–π–æ–Ω—ã, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –∏–º–µ–µ—Ç—Å—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –Ω–µ –º–µ–Ω–µ–µ $30$ –∫–æ–º–ø–∞–Ω–∏–π. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –Ω–æ–≤—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ä–∞–π–æ–Ω–æ–≤, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–≤ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã `vmin` –∏ `vmax` —É —Ñ—É–Ω–∫—Ü–∏–∏ `sns.heatmap`.

```python
df_filtered = df_combined.groupby(['latitude_rounded', 'longitude_rounded']).size()
df_filtered = df_filtered[df_filtered >= 30]

df_combined_filtered = df_combined[
    df_combined.set_index(['latitude_rounded', 'longitude_rounded']).index.isin(df_filtered.index)
]

pivot_table_filtered = pd.pivot_table(df_combined_filtered,
                                      values='average_rating',
                                      index='latitude_rounded',
                                      columns='longitude_rounded',
                                      aggfunc='mean')

plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table_filtered, cmap="YlGnBu", annot=True, fmt='.2f',
            cbar_kws={'label': '–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞'},
            vmin=1, vmax=5, linewidths=0.5)
plt.title('–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –∫–æ–º–ø–∞–Ω–∏–π –ø–æ —Ä–∞–π–æ–Ω–∞–º –õ–∞—Å-–í–µ–≥–∞—Å–∞ (–Ω–µ –º–µ–Ω–µ–µ 30 –∫–æ–º–ø–∞–Ω–∏–π)')
plt.show()
```

–°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º –∏ —Å–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥.

**–í—ã–≤–æ–¥:**
- –ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –æ—Å—Ç–∞–ª–∏—Å—å —Ç–æ–ª—å–∫–æ —Ä–∞–π–æ–Ω—ã —Å –±–æ–ª–µ–µ —Å—Ç–∞–±–∏–ª—å–Ω–æ–π –∏ –Ω–∞–¥–µ–∂–Ω–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–æ–π, –ø–æ—Å–∫–æ–ª—å–∫—É –≤ –Ω–∏—Ö –±–æ–ª—å—à–µ –∫–æ–º–ø–∞–Ω–∏–π, –∏ –∏—Ö —Å—Ä–µ–¥–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ –º–µ–Ω–µ–µ –ø–æ–¥–≤–µ—Ä–∂–µ–Ω—ã –∏—Å–∫–∞–∂–µ–Ω–∏—è–º.
- –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π vmin –∏ vmax –¥–ª—è —Ü–≤–µ—Ç–æ–≤–æ–π –≥–∞–º–º—ã –ø–æ–∑–≤–æ–ª–∏–ª–æ —É–ª—É—á—à–∏—Ç—å –≤–æ—Å–ø—Ä–∏—è—Ç–∏–µ, –æ–±–µ—Å–ø–µ—á–∏–≤ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –º–∞—Å—à—Ç–∞–±—ã –¥–ª—è –≤—Å–µ—Ö —Ä–∞–π–æ–Ω–æ–≤, —á—Ç–æ –¥–µ–ª–∞–µ—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –±–æ–ª–µ–µ –æ–±—ä–µ–∫—Ç–∏–≤–Ω—ã–º.
- –ì—Ä–∞—Ñ–∏–∫ —Å—Ç–∞–ª –±–æ–ª–µ–µ —Ç–æ—á–Ω—ã–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑—É–µ–º—ã–º, —Ç–∞–∫ –∫–∞–∫ –¥–∞–Ω–Ω—ã–µ –æ —Ä–∞–π–æ–Ω–∞—Ö —Å –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–æ–º–ø–∞–Ω–∏–π –±—ã–ª–∏ –∏—Å–∫–ª—é—á–µ–Ω—ã.

–§–∏–ª—å—Ç—Ä–∞—Ü–∏—è —Ä–∞–π–æ–Ω–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∫–æ–º–ø–∞–Ω–∏–π —É–ª—É—á—à–∏–ª–∞ –∫–∞—á–µ—Å—Ç–≤–æ –∞–Ω–∞–ª–∏–∑–∞. –¢–µ–ø–µ—Ä—å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã —Ç–æ–ª—å–∫–æ —Ç–µ —Ä–∞–π–æ–Ω—ã, –≥–¥–µ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –Ω–∞–¥–µ–∂–Ω—ã—Ö –≤—ã–≤–æ–¥–æ–≤. –§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Ü–≤–µ—Ç–æ–≤–æ–π –≥–∞–º–º—ã –ø–æ–∑–≤–æ–ª—è—é—Ç –±–æ–ª–µ–µ —á–µ—Å—Ç–Ω–æ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–∞–∑–ª–∏—á–∏—è –º–µ–∂–¥—É —Ä–∞–π–æ–Ω–∞–º–∏.

#### 3. –†–µ—Å—Ç–æ—Ä–∞–Ω—ã

–ë—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å –∫–æ–º–ø–∞–Ω–∏—é —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–º, –µ—Å–ª–∏ –≤ –ø–æ–ª–µ `categories` *—Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è* —Å–ª–æ–≤–æ `restaurant`. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –≤ –∞–Ω–∞–ª–∏–∑–µ –¥–∞–Ω–Ω—ã—Ö —á–∞—Å—Ç–æ –Ω–µ—Ç —á–µ—Ç–∫–æ–≥–æ —Ñ–æ—Ä–º–∞—Ç–∞ –¥–∞–Ω–Ω—ã—Ö. –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–∞–Ω–Ω–æ–µ —Å–ª–æ–≤–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–∞–ø–∏—Å–∞–Ω–æ –∫–∞–∫ —Å –±–æ–ª—å—à–æ–π –±—É–∫–≤—ã, —Ç–∞–∫ –∏ —Å –º–∞–ª–µ–Ω—å–∫–æ–π; –º–æ–∂–µ—Ç –∫–∞–∫ —Ä–∞–∑–¥–µ–ª—è—Ç—å—Å—è `;`, —Ç–∞–∫ –∏ –Ω–µ —Ä–∞–∑–¥–µ–ª—è—Ç—å—Å—è. –ü—Ä–∏ –≤–æ–∑–Ω–∏–∫–Ω–æ–≤–µ–Ω–∏–∏ –Ω–µ–¥–æ–ø–æ–Ω–∏–º–∞–Ω–∏—è —Å—Ç–æ–∏—Ç –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –¥–∞–Ω–Ω—ã–µ.

–°–æ—Å—Ç–∞–≤—å—Ç–µ —Ç–∞–±–ª–∏—Ü—É, –≤ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –≤—Å–µ—Ö —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞—Ö –≥–æ—Ä–æ–¥–∞ `N`, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –∏–º–µ–µ—Ç—Å—è –Ω–µ –º–µ–Ω–µ–µ $5$ –æ—Ç–∑—ã–≤–æ–≤. –î–∞–ª–µ–µ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ heatmap-–≥—Ä–∞—Ñ–∏–∫ —Ä–∞–π–æ–Ω–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–æ–º—É —Ä–∞–π–æ–Ω—É —Å–æ–ø–æ—Å—Ç–∞–≤—å—Ç–µ —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –ø–æ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞–º —ç—Ç–æ–≥–æ —Ä–∞–π–æ–Ω–∞. –†–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–π—Ç–µ —Ç–æ–ª—å–∫–æ —Ç–µ —Ä–∞–π–æ–Ω—ã, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å –Ω–µ –º–µ–Ω–µ–µ $10$ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤, –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å –Ω–µ –º–µ–Ω–µ–µ $5$ –æ—Ç–∑—ã–≤–æ–≤.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_reviews = pd.read_csv('yelp_review.csv')
df_business = pd.read_csv('yelp_business.csv')

city_name = 'Las Vegas'
df_business_city = df_business[df_business['city'] == city_name].copy()

df_restaurants = df_business_city[df_business_city['categories'].str.contains('restaurant', case=False, na=False)].copy()

df_reviews_filtered = df_reviews[df_reviews['business_id'].isin(df_restaurants['business_id'])].copy()
df_review_counts = df_reviews_filtered.groupby('business_id').size()
valid_restaurants = df_review_counts[df_review_counts >= 5].index

df_valid_restaurants = df_restaurants[df_restaurants['business_id'].isin(valid_restaurants)].copy()

df_valid_restaurants.loc[:, 'latitude_rounded'] = np.round(df_valid_restaurants['latitude'] * 4, decimals=1) * 0.25
df_valid_restaurants.loc[:, 'longitude_rounded'] = np.round(df_valid_restaurants['longitude'] * 4, decimals=1) * 0.25

df_reviews_avg = df_reviews_filtered.groupby('business_id')['stars'].mean()
df_valid_restaurants.loc[:, 'average_rating'] = df_valid_restaurants['business_id'].map(df_reviews_avg)

df_district_counts = df_valid_restaurants.groupby(['latitude_rounded', 'longitude_rounded']).size()
valid_districts = df_district_counts[df_district_counts >= 10].index
df_final = df_valid_restaurants[df_valid_restaurants.set_index(['latitude_rounded', 'longitude_rounded']).index.isin(valid_districts)].copy()

pivot_table = pd.pivot_table(df_final, values='average_rating', index='latitude_rounded', columns='longitude_rounded', aggfunc='mean')

plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap="coolwarm", annot=True, fmt='.2f', cbar_kws={'label': '–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞'}, vmin=1, vmax=5, linewidths=0.5)
plt.title('–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ –ø–æ —Ä–∞–π–æ–Ω–∞–º –õ–∞—Å-–í–µ–≥–∞—Å–∞')
plt.show()
```

–ß–µ–º –ø–æ–ª–µ–∑–Ω—ã –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è —Å–Ω–∏–∑—É –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∑—ã–≤–æ–≤ –¥–ª—è —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ –≤ —Ä–∞–π–æ–Ω–µ?

**–û—Ç–≤–µ—Ç:**
–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è —Å–Ω–∏–∑—É –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∑—ã–≤–æ–≤ –∏ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ –ø–æ–º–æ–≥–∞—é—Ç —É–ª—É—á—à–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –∞–Ω–∞–ª–∏–∑–∞:

- –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è —à—É–º–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö ‚Äì –†–µ—Å—Ç–æ—Ä–∞–Ω—ã —Å 1‚Äì2 –æ—Ç–∑—ã–≤–∞–º–∏ –º–æ–≥—É—Ç –∏–º–µ—Ç—å –Ω–µ—Ä–µ–ø—Ä–µ–∑–µ–Ω—Ç–∞—Ç–∏–≤–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏, –ø–æ—Å–∫–æ–ª—å–∫—É –æ–Ω–∏ –∑–∞–≤–∏—Å—è—Ç –æ—Ç —Å—É–±—ä–µ–∫—Ç–∏–≤–Ω–æ–≥–æ –º–Ω–µ–Ω–∏—è –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤. –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –æ—Ç 5 –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–º–æ–≥–∞–µ—Ç —É—á–∏—Ç—ã–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –±–æ–ª–µ–µ –æ–±—ä–µ–∫—Ç–∏–≤–Ω—ã–µ —Ä–µ–π—Ç–∏–Ω–≥–∏.

–ö–æ—Ç –í–∞—Å–∏–ª–∏–π –∏–∑ –≥–æ—Ä–æ–¥–∞ `N` –æ—á–µ–Ω—å –ø—Ä–∏–¥–∏—Ä—á–∏–≤ –∫ –≤—ã–±–æ—Ä—É —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞. –û–Ω –¥–æ–≤–µ—Ä—è–µ—Ç —Ç–æ–ª—å–∫–æ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞–º —Å –≤—ã—Å–æ–∫–∏–º —Ä–µ–π—Ç–∏–Ω–≥–æ–º, –∫–æ—Ç–æ—Ä—ã–π –æ—Å–Ω–æ–≤—ã–≤–∞–µ—Ç—Å—è –Ω–∞ –±–æ–ª—å—à–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –æ—Ç–∑—ã–≤–æ–≤. –ù–∞–ø–µ—á–∞—Ç–∞–π—Ç–µ –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é $10$ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞—Ö —Å —Å–∞–º—ã–º –±–æ–ª—å—à–∏–º —Ä–µ–π—Ç–∏–Ω–≥–æ–º –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è —Ä–µ–π—Ç–∏–Ω–≥–∞. –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ —ç—Ç–∏—Ö —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –Ω–µ –º–µ–Ω–µ–µ $50$ –æ—Ç–∑—ã–≤–æ–≤. –ü–æ –∫–∞–∂–¥–æ–º—É —Ä–µ—Å—Ç–æ—Ä–∞–Ω—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ —Å–ª–µ–¥—É—é—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏: –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞, —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∑—ã–≤–æ–≤, –≥–µ–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã, –∫–∞—Ç–µ–≥–æ—Ä–∏–∏.

```python
df_reviews_counts = df_reviews_filtered.groupby('business_id').size().reset_index(name='number_of_reviews')
df_valid_restaurants = df_restaurants.merge(df_reviews_counts, on='business_id')
df_valid_restaurants = df_valid_restaurants[df_valid_restaurants['number_of_reviews'] >= 50]

df_reviews_avg = df_reviews_filtered.groupby('business_id')['stars'].mean().reset_index(name='average_rating')
df_valid_restaurants = df_valid_restaurants.merge(df_reviews_avg, on='business_id')

df_top_restaurants = df_valid_restaurants.sort_values(by='average_rating', ascending=False).head(10)

df_top_restaurants = df_top_restaurants[['name', 'average_rating', 'number_of_reviews', 'latitude', 'longitude', 'categories']]
print(df_top_restaurants)
```

–ù–∞–Ω–µ—Å–∏—Ç–µ –Ω–∞ –∫–∞—Ä—Ç—É –≤—Å–µ —Ä–µ—Å—Ç–æ—Ä–∞–Ω—ã —Å–æ —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –Ω–µ –º–µ–Ω–µ–µ $4.7$, –∫–æ—Ç–æ—Ä–∞—è –ø–æ—Å—á–∏—Ç–∞–Ω–∞ –ø–æ –Ω–µ –º–µ–Ω–µ–µ $50$ –æ—Ç–∑—ã–≤–∞–º. –û—Ç–¥–µ–ª—å–Ω—ã–º —Ü–≤–µ—Ç–æ–º –æ—Ç–º–µ—Ç—å—Ç–µ 10 —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã –ø–æ–ª—É—á–∏–ª–∏ —Ä–∞–Ω–µ–µ.

```python
import folium
from folium.plugins import MarkerCluster

df_best_restaurants = df_valid_restaurants[
    (df_valid_restaurants['average_rating'] >= 4.7) & (df_valid_restaurants['number_of_reviews'] >= 50)
]

las_vegas_map = folium.Map(location=[36.1699, -115.1398], zoom_start=13, tiles='OpenStreetMap')

marker_cluster = MarkerCluster().add_to(las_vegas_map)

for _, row in df_best_restaurants.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"{row['name']}<br>–ö–∞—Ç–µ–≥–æ—Ä–∏–∏: {row['categories']}<br>–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞: {row['average_rating']}<br>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∑—ã–≤–æ–≤: {row['number_of_reviews']}",
        icon=folium.Icon(color='blue')  # –ì–æ–ª—É–±–æ–π —Ü–≤–µ—Ç –¥–ª—è –≤—Å–µ—Ö —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ 4.7+
    ).add_to(marker_cluster)

for _, row in df_top_restaurants.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"–¢–û–ü-10! {row['name']}<br>–ö–∞—Ç–µ–≥–æ—Ä–∏–∏: {row['categories']}<br>–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞: {row['average_rating']}<br>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–∑—ã–≤–æ–≤: {row['number_of_reviews']}",
        icon=folium.Icon(color='red')  # –ö—Ä–∞—Å–Ω—ã–π —Ü–≤–µ—Ç –¥–ª—è –¢–û–ü-10
    ).add_to(las_vegas_map)

las_vegas_map.save("las_vegas_best_restaurants_map.html")
print("–ö–∞—Ä—Ç–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –∫–∞–∫ 'las_vegas_best_restaurants_map.html'")
```
